```python
import csv
import requests
import time
from bs4 import BeautifulSoup

def blog_crawling(page=1):
    #미국여행지추천과 관련된 url을 복사하여 넣는다.
    url ='https://search.naver.com/search.naver?where=post&sm=tab_jum&query=%EB%AF%B8%EA%B5%AD%EC%97%AC%ED%96%89%EC%A7%80%EC%B6%94%EC%B2%9C'.format(page)

    #요청을 하고 응답이 제대로 오는지 확인한다.
    response = requests.get(url)
    #도착한 응답을 beautiful Soup로 파싱한다.
    soup = BeautifulSoup(response.text, 'html.parser')

    blog_post_list = []
    #블로그의 제목과 내용을 크롤링한다.

    for links in soup.select('li.sh_blog_top > dl'):
        title = links.select('dt > a')#dt태그 밑에 a태그=제목
        content = links.select('dd.sh_blog_passage')#dd태그의 sh_blog_passage클래스=내용

        # select로 찾은 결과는 리스트로 표시되기에 [0]을 씀. 
        title = title[0].get('title')

        content = content[0].text


        #블로그 데이터를 사전으로 만들어줌.
        blog_post = {'title': title, 'content': content}

        blog_post_list.append(blog_post)

    return blog_post_list

def save_data(blog_post):
    #추출한 데이터를 엑셀에 저장
    keys = blog_post[0].keys()
    with open('blog_crawling.csv', 'w') as file:
        writer = csv.DictWriter(file, keys)
        writer.writeheader()
        writer.writerows(blog_post)
    #100개 블로그의 데이터 저장
    blog_post_list = []
    for i in range(1, 100, 10):
        blog_post_list.extend( blog_crawling(page=i) )
        time.sleep(2)

    save_data(blog_post_list)
```
